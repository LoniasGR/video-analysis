### Άσκηση 1: Ζωγραφίστε το γράμμα H
Ορίστε μια συνάρτηση που λαμβάνει ως είσοδο μια εικόνα RGB και ένα ζεύγος συντεταγμένων (σειρά, στήλη) και επιστρέφει ένα αντίγραφο της αρχικής εικόνας με ένα πράσινο γράμμα H επικαλυμμένο σε συγκεκριμένες διαστάσεις της εικόνας.

Οι συντεταγμένες δείχνουν την επάνω αριστερή γωνία του Η. Το πάχος του γράμματος θα πρέπει να είναι 3 εικονοστοιχείων και το ύψος και το πλάτος του να είναι 24 pixel.

Ξεκινήστε με το ακόλουθο πρότυπο:

```python
def draw_H(image, coords, color=(0, 255, 0)):
    out = image.copy()
    x_start = coords[0]
    y_start = coords[1]
    out[x_start:x_start+24, y_start:y_start+3] = list(color)
    out[x_start+11:x_start+13, y_start:y_start+24] = list(color)
    out[x_start:x_start+24, y_start+21:y_start+24] = list(color)
    
    return out
```

Δοκιμάστε το πρόγραμμά σας με το ακόλουθο κώδικα:

```python
cat = data.chelsea()
cat_H = draw_H(cat, (50, -50))
plt.imshow(cat_H);
```



### Άσκηση 2: Απεικόνιση καναλιών RGB 
Εμφανίστε τα διαφορετικά χρωματικά κανάλια χρώματος της εικόνας(κάθε μία ως εικόνα με γκρι κλίμακα). 

Ξεκινήστε με το ακόλουθο πρότυπο:

```python
# --- read in the image ---

image = plt.imread('balloon.jpg')

# --- assign each color channel to a different variable ---

r = image[..., 0]
g = image[..., 1]
b = image[..., 2]

# --- display the image and r, g, b channels ---

f, axes = plt.subplots(1, 4, figsize=(16, 5))

for ax in axes:
    ax.axis('off')

(ax_r, ax_g, ax_b, ax_color) = axes
    
ax_r.imshow(r, cmap='gray')
ax_r.set_title('red channel')

ax_g.imshow(g, cmap='gray')
ax_g.set_title('green channel')

ax_b.imshow(b, cmap='gray')
ax_b.set_title('blue channel')

# --- Here, we stack the R, G, and B layers again
#     to form a color image ---
ax_color.imshow(np.stack([r, g, b], axis=2))
ax_color.set_title('all channels');
```

### Άσκηση 3: Μετατροπή σε κλίμακα του γκρι ("μαύρο και άσπρο") 
Η σχετική φωτεινότητα μιας εικόνας είναι η ένταση του φωτός που προέρχεται από κάθε σημείο. Τα διαφορετικά χρώματα συμβάλλουν διαφορετικά στη φωτεινότητα: είναι πολύ δύσκολο να έχουμε για παράδειγμα ένα φωτεινό, καθαρό μπλε χρώμα. Έτσι, ξεκινώντας από μια εικόνα RGB, η φωτεινότητα της δίνεται από τον τύπο: 

$$ Υ = 0,2126R + 0,7152G + 0,0722Β $$


α) Χρησιμοποιήστε τον πολλαπλασιασμό πινάκων, για να μετατρέψετε μια εικόνα RGB σε μια εικόνα φωτεινότητας της κλίμακας του γκρι σύμφωνα με τον παραπάνω τύπο.

β) Αλλάξτε τους συντελεστές στο 1/3 (δηλαδή, πάρτε τον μέσο όρο του κόκκινου, πράσινου και μπλε καναλιού, και συγκρίνεται αυτή η προσέγγιση με το rgb2gray.

γ) Συγκρίνετε τα αποτελέσματά σας από το α) και β) με αυτά που έχετε πάρει με τη συνάρτηση skimage.color.rgb2gray.ray.

```python
from skimage import color, img_as_float, io


image = img_as_float(io.imread('balloon.jpg'))

r = image[..., 0]
g = image[..., 1]
b = image[..., 2]

gray = color.rgb2gray(image)
my_gray = 1/3 * r + 1/3 * g + 1/3 * b
Y = 0.2126 * r + 0.7152 * g + 0.0722 * b

# --- display the results ---

f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(10, 6))

ax0.imshow(gray, cmap='gray')
ax0.set_title('skimage.color.rgb2gray')

ax1.imshow(my_gray, cmap='gray')
ax1.set_title('my rgb2gray')

ax2.imshow(Y, cmap='gray')
ax2.set_title('my Y')
```
